```python
def create_columns(list, num_columns):
    string = ""
    section_count = 1
    for i, item in enumerate(list):
        # first part of string
        if string == "":
            string += f"{item['Label']} & {item['Value']}"
        # if ending should be added
        elif i%num_columns == 0:
            section_count += 1
            string += f" \\\\ {item['Label']} & {item['Value']}"
        else:
            string += f" & {item['Label']} & {item['Value']}"

    for i in range(0, section_count*num_columns - len(list)):
        string += " & &"

    print(string)
        

create_columns(list1, 1)
create_columns(list1, 2)
create_columns(list1, 3)

create_columns(list2, 1)
create_columns(list2, 2)
create_columns(list2, 3)

create_columns(list3, 1)
create_columns(list3, 2)
create_columns(list3, 3)

create_columns(list4, 1)
create_columns(list4, 2)
create_columns(list4, 3)
```
